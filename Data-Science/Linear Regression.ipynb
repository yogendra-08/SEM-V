```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'Rent.xlsx')
df
```

```python
df.shape
```

### Step 1: Apply Preprocessing Steps

```python
df.isnull().sum()
```

### Step 2: Separate independent and dependent variable

```python
x = df.drop('rent',axis=1)
x
```

```python
y = df['rent']
y
```

### Step 3: Splitting data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

### Step 4: Algorithm Implementation

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
LR
```

### Step 5: Model Train

```python
LR.fit(x_train, y_train)
```

### Step 6: Model Predict

```python
y_pred = LR.predict(x_test)
y.36419_pred
```

```python
y_test
```

### Step 7: Error

```python
error = y_test - y_pred
error
```

### Step 8: Check Score

#### Training Score

```python
LR.score(x_train,y_train)
```

#### Testing Score

```python
LR.score(x_test,y_pred)
```
