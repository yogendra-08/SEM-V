```python
from nltk.steam import parter.steammer 
from nltk.tokenize import sent_tokenize, word_tokenize

ps = parter.steammer()

wordss = ["pythoner", "pythoning", "pythoned", "pythonly"]

for w in wordss:
    print(ps.stem(w))

```

```python
import nltk 
nltk.download('punkt')

text = "it is very important to be pythonly while you are pythoning with python. " \
"All pythoners have pythoned poorly at least once."

word = word_tokenize(text)

for w in word:
    print(ps.stem(w))

```

```python
from nltk.tokenize import sent_tokenize, word_tokenize

text = "Hello Mr. Smith, how are you doing today? The weather is great, and Python is awesome. The sky is pinkish-blue. You shouldn't eat cardboard."

print(sent_tokenize(text))
print(word_tokenize(text))
```
