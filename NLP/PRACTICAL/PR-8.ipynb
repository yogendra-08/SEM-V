```python
import nltk
import random
from nltk.corpus import names

# Download the 'names' corpus (run once)
nltk.download('names')

# Create labeled (name, gender) pairs
labeled_names = ([(name, 'male') for name in names.words('male.txt')] +
                 [(name, 'female') for name in names.words('female.txt')])

# Shuffle the data randomly
random.shuffle(labeled_names)

# Feature extractor function
def gender_features(word):
    return {'last_letter': word[-1].lower()}

# Create feature sets: (features, label)
featuresets = [(gender_features(n), g) for (n, g) in labeled_names]

# Split into training and testing data
train_set, test_set = featuresets[500:], featuresets[:500]

# Train Naive Bayes classifier
classifier = nltk.NaiveBayesClassifier.train(train_set)

# Predict gender for name 'Neo'
print("Prediction for Neo:", classifier.classify(gender_features('Neo')))

# Calculate accuracy on test data
print("Accuracy:", nltk.classify.accuracy(classifier, test_set))

# Show the most informative features
classifier.show_most_informative_features(5)
```
